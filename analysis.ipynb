## importing libraries and dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("LoanApprovalPrediction.csv")
```

```python
data.head(5)
```

```python
# summary of categorical variables

obj = (data.dtypes == 'object')
print("Categorical variables:",len(list(obj[obj].index)))
```

```python
# Dropping Loan_ID column since it is a unique identifier.
data.drop(['Loan_ID'],axis=1,inplace=True)
```

```python
# summary of numerical variables

obj = (data.dtypes == 'object')
object_cols = list(obj[obj].index)
plt.figure(figsize=(18,36))
index = 1

for col in object_cols:
    y = data[col].value_counts()
    plt.subplot(11,4,index)
    plt.xticks(rotation=90)
    sns.barplot(x=list(y.index), y=y)
    index +=1
```

```python
# Import label encoder for categorical variables that are in string format to convert them into numeric format
from sklearn import preprocessing
	
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
obj = (data.dtypes == 'object')
for col in list(obj[obj].index):
    data[col] = label_encoder.fit_transform(data[col])
```

```python
# To find the number of columns with
# datatype==object after label encoding
obj = (data.dtypes == 'object')
print("Categorical variables:",len(list(obj[obj].index)))
```

```python
# Heat map to find the correlation between the variables

plt.figure(figsize=(12,6))

sns.heatmap(data.corr(),cmap='BrBG',fmt='.2f',
			linewidths=2,annot=True)
```

```python
# Plotting the distribution of the variables 

sns.catplot(x="Gender", y="Married",
			hue="Loan_Status",
			kind="bar",
			data=data)
```

```python
# looking for null values

for col in data.columns:
    data[col] = data[col].fillna(data[col].mean())
	
data.isna().sum()
```

## Machine Learning Model Building.

```python
# Splitting the data into train and test

from sklearn.model_selection import train_test_split

X = data.drop(['Loan_Status'],axis=1)
Y = data['Loan_Status']
X.shape,Y.shape

X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
													test_size=0.4,
													random_state=1)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

# Model Training and Evaluation

```python
# Importing the models, fitting them on the training set and making predictions on the training set

from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors=3)
rfc = RandomForestClassifier(n_estimators = 7,
							criterion = 'entropy',
							random_state =7)
svc = SVC()
lc = LogisticRegression()

# making predictions on the training set
for clf in (rfc, knn, svc,lc):
	clf.fit(X_train, Y_train)
	Y_pred = clf.predict(X_train)
	print("Accuracy score of ",
		clf.__class__.__name__,
		"=",100*metrics.accuracy_score(Y_train,
										Y_pred))
```

```python
# making predictions on the testing set to check their accuracy

for clf in (rfc, knn, svc,lc):
	clf.fit(X_train, Y_train)
	Y_pred = clf.predict(X_test)
	print("Accuracy score of ",
		clf.__class__.__name__,"=",
		100*metrics.accuracy_score(Y_test,
									Y_pred))
```

```python
# Plotting the confusion matrix for the best model

from sklearn.metrics import confusion_matrix

rfc.fit(X_train, Y_train)
Y_pred = rfc.predict(X_test)
cm = confusion_matrix(Y_test, Y_pred)
sns.heatmap(cm, annot=True, fmt="d")

```

```python
# Saving the linear regression model

import pickle
pickle.dump(rfc, open('model.pkl','wb'))

```
